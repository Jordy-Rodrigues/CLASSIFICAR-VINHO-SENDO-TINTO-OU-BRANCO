**CLASSIFICAR UMA BASE DE DADOS DE VINHO PELA SUA ÁCIDES COMO SENDO VINHO TINTO OU VINHO BRANCO**

```python
#IMPORTAR BIBLIOTECA BANCDAS E LER A BASE DE DADOS
import pandas as pd
arquivo=pd.read_csv("/content/drive/MyDrive/wine_dataset.csv")
```

```python
# MOSTRAR A TABELA
arquivo
```

```python
# TRANSFORMAR A COLUNA 'Style' DE TEXTO PARA NUMÉRICA
arquivo['style']= arquivo['style'].replace('red', 0)
arquivo['style']= arquivo['style'].replace('white', 1)
```

```python
# MOSTRAR A TABELA
arquivo
```

```python
# SEPARAR AS VARIAVEIS ENTRE PREDITORAS E VARIAVEIS ALVO
y=arquivo['style']
x=arquivo.drop('style', axis=1)
```

```python
print(y)
```

```python
print(x)
```

PERAR OS DADOS PARA O MODELO 

```python
# IMPORTAR O MODELO DA BIBLIOTECA SKLEARN
from sklearn.model_selection import train_test_split
```

```python
# CRIAR UM CONJUNTO DE DADOS DE TREINO E TESTE
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.3)
```

```python
# VER O TOTAL DE LINHAS E COLUNAS DO DATASET
arquivo.shape
```

```python
# BASE DE DADOS SEPARADOS PARA TREINO
x_treino.shape
```

```python
y_treino.shape
```

CRIAR O MODELO DE MACHINE LEARNING

```python
# IMPORTAR A BIBLIOTECA CLASSIFICADOR DE ÁRVORES EXTRA
from sklearn.ensemble import ExtraTreesClassifier
modelo= ExtraTreesClassifier()
modelo.fit (x_treino, y_treino)
```

```python
# CONFERIR O RESULTADO DO TREINO
resultado= modelo.score (x_teste, y_teste)

print('acurácia: ', resultado)
```

```python
# 5 DADOS ALEATÓRIOS DE TESTE PARA O MODELO TENTAR PREVER
y_teste [400:405]
```

```python
# MODELO VAI TENTAR PREVER OS 5 DADOS DE TESTE QUE FORAM SEPARADOS A CIMA
previsoes= modelo.predict(x_teste [400:405])
previsoes
```
